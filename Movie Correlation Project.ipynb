```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create


# Read the data
df = pd.read_csv(r'movies.csv') # is r optional?
```

```python
df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for columns

df.dtypes
```

```python
# Replacing NULL with 0
df = df.fillna(0)
df

df.loc[df["company"] == 0, "company"] = "Unknown"
```

```python
# Changing data type of columns to remove ending .0's
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head()
```

```python
# Create correct Year column

# df['yearcorrect'] = df['released'].str.rsplit(',').str[-1]
df['yearcorrect'] = df['released'].str.rsplit('(').str[0].str.rsplit(',').str[-1]

df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop duplicates showing unique values only
# Also checking for quality of data

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.head()
```

```python
# Budget High Correlation
# Company High Correlation
```

```python
# Scatter plot with budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')
plt.show()
```

```python
df.head()
```

```python
# Creating a regression block with Seaborn
# Plot budget vs gross

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Looking at the correlation

df.corr(method='pearson') #Pearson(default), kendall and spearman are available for deterniming correlation
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Looking at Company column

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
        
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_matrix = df_numerized.corr()
corr_pairs = correlation_matrix.unstack()
corr_pairs
```

```python
# Quicker way than above
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr

# Votes and budget have the highest correlation to gross earnings
# Company has low correlation
```
